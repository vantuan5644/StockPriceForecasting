```python
import vnquant.DataLoader as web
```

```python
def get_stock_price_history(name: str, start_date: str, end_date: str):
    loader = web.DataLoader(name, start=start_date, end=end_date)
    data = loader.download()
    return data
```

```python
data = get_stock_price_history('FPT', '2018-01-01', '2020-12-27')
```
